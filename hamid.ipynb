```python
import numpy as np
flow_local = np.load('flow_local_out.npy')
corrected_flow = np.load('corrected_flow_out.npy')
```

```python
import matplotlib.pyplot as plt


erreur_scale = (corrected_flow[:,1] - flow_local[:,1])
t= np.linspace(0,1000,1000)*10**(-6)
# Tracé de l'erreur de mise à l'échelle
plt.figure(figsize=(10, 5))
plt.plot(t,erreur_scale[0:1000],label='correction angulaire')
plt.xlabel('Temps (sec)')
plt.title('correction angulaire (ARMS - EDL)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


erreur_scale = (corrected_flow[:,0] - flow_local[:,0])
t= np.linspace(0,1000,1000)*10**(-6)
# Tracé de l'erreur de mise à l'échelle
plt.figure(figsize=(10, 5))
plt.plot(t,erreur_scale[0:1000],label='correction magnitude')
plt.xlabel('Temps (sec)')
plt.title('correction magnitude (ARMS - EDL)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Créer un diagramme polaire
ax = plt.subplot(111, polar=True)
# Tracer les données pour ARMS
ax.bar(flow_local[0:50,1], flow_local[0:50,0], width=0.2, alpha=0.5, label='ARMS-Flow')
ax.bar(corrected_flow[0:50,1], corrected_flow[0:50,0], width=0.2, alpha=0.5, label='EDL-Flow')
# Tracer les données pour EDL
ax.set_theta_zero_location('E') 
ax.set_theta_direction(1)  

# Ajouter des légendes, des titres, etc.
ax.set_title("Direction distributions")
ax.legend()

plt.show()
```
