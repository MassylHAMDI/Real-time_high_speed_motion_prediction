```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio

src_serie = sio.loadmat('datamat.mat')
x  = src_serie["x"].reshape(-1,)       
y  = src_serie["y"].reshape(-1,)       
ts = src_serie["ts"].reshape(-1,) 
p  = src_serie["p"].reshape(-1,)       
events = np.vstack((x, y, ts, np.ones(ts.shape))).transpose() 
max_x, max_y, max_t, _ = np.max(events, axis = 0)
min_x, min_y, min_t, _ = np.min(events, axis = 0)
```

```python
min_range, max_range = (0,5000)
show_polarity = False
fig = plt.figure(figsize = (8,8)) 
ax  = fig.add_subplot(111, projection = '3d')
# select the events
events_ = events[min_range : max_range, :3].T
if show_polarity : 
    # Show the events pol
    pol_ = p[min_range : max_range]
    pol_color = np.where( pol_ == 1, "blue", "red")
    ax.scatter(*events_, color = pol_color, s = 8)
else :
    ax.scatter(*events_, s = 8)
ax.set_xlabel(r'x (px)')
ax.set_ylabel(r'y (px)')
ax.set_zlabel(r'ts (µs)')
plt.show()
```
